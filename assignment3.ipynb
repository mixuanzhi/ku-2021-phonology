```python
import pandas as pd
df = pd.read_csv('/yourgoodpants.csv')
print(df)
```

```python
df.loc[df.Time_s == "--undefined--", "Time_s"] =None
df.loc[df.F0_Hz == "--undefined--", "F0_Hz"] =None
print(df)
```

```python
time = df.Time_s
f0 = df.F0_Hz
import matplotlib.pyplot as plt

plt.plot(time, f0, 'ro', label='F0 (Hz)')
plt.xlabel('time(sec)')
plt.ylabel('pitch(Hz)')
plt.legend(loc='best', fontsize=14, frameon=True, shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/yourgoodpants_intensity.csv')
print(df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('/yourgoodpants_formants.csv')
print(df)
```

```python
time = df.Time_s
plt.xlabel('time(sec)')
plt.ylabel('Frequency(Hz)')
F1_Hz = df.F1_Hz
F2_Hz = df.F2_Hz
F3_Hz = df.F3_Hz
F4_Hz = df.F4_Hz

plt.plot(time, F1_Hz, 'r-', label='F1')
plt.plot(time, F2_Hz, 'b-', label='F2')
plt.plot(time, F3_Hz, 'g-', label='F3')
plt.plot(time, F4_Hz, 'y-', label='F4')
plt.legend(loc='best', ncol=1, fontsize=14, frameon=False, shadow=True)
plt.show()
```
