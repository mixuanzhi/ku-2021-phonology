```python
import numpy as np
import matplotlib.pyplot as plt
import IPython.display as ipd
```

```python
def getFrequency(note):
  A4=440
  notes = ['A', 'A#', 'B', 'C', 'C#', 'D', 'D#', 'E', 'F', 'F#', 'G', 'G#' ]

  octave = int(note[2]) if len(note)  == 3 else int(note[1])

  keyNumber = notes.index(note[0:-1]);

  if (keyNumber < 3) :
    keyNumber = keyNumber + 12 + ((octave - 1) * 12) + 1;
  else:
    keyNumber = keyNumber + ((octave - 1) * 12) + 1;
  return A4 * 2** ((keyNumber- 49) / 12) 
```

```python
amp = 1; sr = 10000; phase = 0

F1 = getFrequency('F1')
F2 = getFrequency('F2')
F4 = getFrequency('F4')
F5 = getFrequency('F5')
F3 = getFrequency('F3')
```

```python
# Jingle Bells

notes = [F3, F3, F3, F3, F3, F3, F3, F5, F1, F2, F3, 0, \
         F4, F4, F4, F4, F4, F3, F3, F3, F3, F3, F2, F2, F1, F2, F5]

durs = [0.25, 0.25, 0.5, 0.25, 0.25, 0.5, 0.25, 0.375, 0.125, 0.25, 0.5, 0.5, \
        0.25, 0.25, 0.375, 0.125, 0.25, 0.25, 0.25, 0.125, 0.125, 0.25, 0.25, 0.25, 0.25, 0.5, 0.5]
```

```python
def phasor_sine(amp, sr, dur, freq, phase):
  t = np.arange(1, sr * dur+1)/sr
  theta = t * 2*np.pi * freq + phase
  s = amp * np.sin(theta)
  return s, t, theta
```

```python
s_all = np.empty(0)

for note, dur in zip(notes, durs):
  s, t, th= phasor_sine(amp, sr, dur, note, phase)
  s_all = np.concatenate((s_all, s))
```

```python
ipd.Audio(s_all, rate=sr)
```
